```python
x = np.linspace(0,2*np.pi,num=100)
x
```

```python
siny = np.sin(x)
siny
```

```python
plt.plot(x,siny)
```

```python
cosy = np.cos(x)
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
plt.figure(figsize=(10,8))#画布大小
plt.plot(x, siny, color='r', marker = '*', linestyle = '--', label ='sin(x)')
plt.plot(x, cosy, color='g', marker = 'o', linestyle = '-.', label ='cos(x)')
#plt.plot(x, cosy, color='g', marker = '+')
plt.legend(loc=3)
plt.xlabel('频率 hz', fontsize=15)
plt.ylabel('电压 v', fontsize=15)
plt.title("欢迎使用matplotlib绘图工具", fontsize=15)
plt.show()
```

```python
#饼状图
x = np.random.randint(3000,500000,10)
plt.figure(figsize=(10,8))
import string
names = [f'department {v}' for v in string.ascii_uppercase[:10]]
explodes=[0.0 for _ in range(len(names))]
explodes[0]=0.3
plt.pie(x, labels=names, explode=explodes, shadow=True, autopct='%1.2f%%')
plt.legend()
plt.savefig('img1.png')
plt.show() #结束
```

```python
#直方图
np.random.seed(666)
mu, sigma=100, 15
x = mu+sigma*np.random.randn(100000) # normal
#x = np.random.normal(mu, sigma, 100000)
plt.figure(figsize=(8,6))
plt.hist(x, bins=50, density=True, facecolor='g',alpha=0.75)
plt.xlabel("IQ")
plt.xlabel("频数")
plt.xlim([40,160])
plt.ylim([0, 0.03])
plt.text(60, 0.025, r'$\mu = 100, \sigma = 15$', fontsize=15)
plt.grid()
plt.show()
```
