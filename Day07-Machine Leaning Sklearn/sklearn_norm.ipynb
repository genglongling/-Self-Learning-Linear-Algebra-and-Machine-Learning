```python
from sklearn import datasets
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
iris = datasets.load_iris()
X=iris.data
y=iris.target
#划分数据集
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)
(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# 1. get mean, std
standardScaler = StandardScaler()
standardScaler.fit(X_train)
```

```python
standardScaler.mean_ #mean
```

```python
standardScaler.scale_ #std
```

```python
# 2. get transform result
X_train_stand = standardScaler.transform(X_train)
X_train_stand
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train_stand, y_train)
```

```python
# 3. do the same for X_test
standardScaler.fit(X_test)
X_test_stand=standardScaler.transform(X_test)
```

```python
knn_clf.score(X_test_stand, y_test)
```
